# Generate UR strings of length 2m such that:

## the number of right moves is always less than or equal to the triangular number corresponding to the number of up moves

## if removeCatalans == True, gets rid of all paths that satisfy y>=x at all steps

```python
def generate_binary_strings(m,removeCatalans):
    bit_count = 2*m-1
    binary_strings = []
    def genbin(n, bs=''):
        if len(bs) == n:
            binary_strings.append(bs)
        else:
            genbin(n, bs + '0')
            genbin(n, bs + '1')


    genbin(bit_count)
    

    for string in binary_strings:
        curValue = 0
        badTriangles = False
        isCatalan = True
        UCount = 0
        RCount = 0
        for char in string:
            curValue += int(char)
            if char =='1':
                UCount += 1
            else:
                RCount += 1
            if RCount > UCount*(UCount+1)/2:
                badTriangles = True
            if RCount > UCount:
                isCatalan = False
        binary_strings[binary_strings.index(string)] = (string + '0',badTriangles,curValue,isCatalan)

    if removeCatalans == False:
        validStrings = [x[0] for x in binary_strings if x[0][0]=='1' and x[1]==False and x[2]==m]
    else:
        validStrings = [x[0] for x in binary_strings if x[0][0]=='1' and x[1]==False and x[2]==m and x[3]==False]
    # print(validStrings)
    for string in validStrings:
        newString = ''
        for char in string:
            if char == '1':
                newString = newString + 'U'
            else:
                newString = newString + 'R'
        validStrings[validStrings.index(string)] = newString
    return validStrings
```

## check if the sets match

```python
def find_differences(l1,l2):
    print(' ')
    print('In list 1 but not 2:')
    A = list(set(l1).difference(l2))
    if A != []:
        for string in A:
            print(string)
    else:
        print('none')
    print(' ')
    print('In list 2 but not 1:')
    B = list(set(l2).difference(l1))
    if B != []:
        for string in B:
            print(string)
    else:
        print('none')
    print(' ')
```

## set up code for OEIS A127496 and specific code for A127498 (which has a(n)=A127496(n+1,n))

```python
def get_A127496(n,k):
    if n<0 or k<0 or k>n*(n+1)/2:
        res = 0
    elif k==0:
        res = 1
    elif k<=n*(n-1)/2:
        res = get_A127496(n,k-1) + get_A127496(n-1,k)
    else:
        res = get_A127496(n,k-1)
    return res

def get_number(m):
    return get_A127496(m,m-1)
```
